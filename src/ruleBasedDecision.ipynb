# Règles pour la décisions
Pour aider à la décision de l'intelligence artificielle, un certain nombre de règles sont établient pour permettre d'être plus précis dans les cas simples.<br>
Un certain nombre de pics sont référencés, si un de ces pic est supérieur à un seuil, le chromatogramme ne sera pas considéré comme normale, et une justification (nom de la molécule ou nom de la maladie) pourra être affichée. <br>
Ce résultat sera là en complément de l'analyse par intelligence artificielle.

```python
import pandas as pd
import tools

pd.options.plotting.backend = "plotly"
PATH = '../data/all-data/'
```

```python
data = tools.readAllData(PATH)
```

```python
for i in data[:4]:
    print("\nFichier : ",i.name)
    fig = i.df.plot()
    fig.show()
    print(i.printProblems())
```

Vérification que les problèmes sont détectés

```python
n = 0
for i in data:
    result = i.problemsDescription()
    
    if (result != "") :
        n+=1
print('nombre de données considérés anormales : ', n/len(data) * 100, ' %')
print(n)
```
